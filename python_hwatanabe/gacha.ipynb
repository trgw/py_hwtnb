```
import random
```

```
def gacha(n):
  cd = 0
  posters = []
  while len(set(posters)) < n:
    posters.append(random.randint(1, n))
    cd += 1
  return cd
```

```
trial = 100
cd = 0
N = 44
for _ in range(trial):
  cd += gacha(N)
print(cd/trial)
```

```
def p(n):
  r = 0.0
  for i in range(n):
    r += n/(i+1)
  return r

p(44)
```
