```
from PIL import Image, ImageDraw
from IPython.display import display
```

```
def newton(x):
  for _ in range(20):
    x = x - (x**3-1)/(3*x**2)
  return x
```

```
def plot(draw, s):
  hs = s/2
  red = (255, 0, 0)
  green = (0, 255, 0)
  blue = (0, 0, 255)
  for x in range(s):
    for y in range(s):
      z = complex(x-hs+0.5, y-hs+0.5)/s*4
      z = newton(z)
      if  z.real >= 0:
        c = red
      elif z.real < 0 and z.imag >= 0:
        c = green
      else:
        c = blue
      draw.rectangle([x, y, x+1, y+1], fill=c)
```

```
size = 512
im = Image.new("RGB", (size, size))
draw = ImageDraw.Draw(im)
plot(draw, size)
display(im)
```
