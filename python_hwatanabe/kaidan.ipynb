```
from graphviz import Digraph
import IPython
```

```
def kaidan(n):
  #終端条件
  if n == 1:
    return 1
  if n == 2:
    return 2
  #再帰部分
  return kaidan(n-1) + kaidan(n-2)
```

```
kaidan(3)
```

```
kaidan(4)
```

```
kaidan(5)
```

```
kaidan(10)
```

```
def kaidan_g(n, g, nodes, parent=None):
  index = str(len(nodes))
  nodes.append(index)
  g.node(index, str(n))

  if parent is not None:
    g.edge(index, parent)

  if n in (1, 2):
    return
  kaidan_g(n-1, g, nodes, index)
  kaidan_g(n-2, g, nodes, index)
```

```
n = 8
graph = Digraph(format='png')
kaidan_g(n, graph, [])
graph.graph_attr.update(size='10,10')
graph.render('test')
IPython.display.Image('test.png')
```

```
n = 8
graph = Digraph(format='png')
kaidan_g(n, graph, [])
graph.graph_attr.update(size='5,5')
graph.render('test')
IPython.display.Image('test.png')
```
