```
import matplotlib.pyplot as plt
from scipy import linalg
import numpy as np
```

```
N = 32
V = np.array([-5.0 if i in range(N//4, 3*N//4) else 0 for i in range(N)])
H = np.zeros((N, N))
for i in range(N):
  i1 = (i + 1) % N
  i2 = (i - 1 + N) % N
  H[i][i] = 2.0 + V[i]
  H[i][i1] = -1
  H[i][i2] = -1
```

```
w, v = linalg.eigh(H)
i0 = np.argmin(w)
print(w[i0])
```

```
v = v.transpose()[i0]
v = v * v
plt.plot(v*20+w[i0])
plt.plot(V)
```
